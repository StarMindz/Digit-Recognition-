#Build Model for Optical character Recognition using MNIST dataset in tensorflow


```python
#import mnist module from tensorflow

from tensorflow.keras.datasets import mnist
import matplotlib.pyplot as plot
```

```python
data=mnist.load_data
```

```python
(X_train, y_train), (X_test, y_test)=data()
```

```python
print(X_train.shape,"\n",
y_train.shape,"\n",
X_test.shape, "\n",
y_test.shape)
```

```python
#To see any of the images in the X_train dataset,  use the imshow function in matplotlib.pyplot
plot.title("Image Label: " +str(y_train[0]))          #image label
plot.imshow(X_train[0])    #image array data
plot.axis("off")
```

```python
#Do this for the first 10 image data

for index in range(10):
  plot.subplot(2,5,index+1)
  plot.imshow(X_train[index])
  plot.axis("off")
  plot.title("Label : "+ str(y_train[index]))
plot.show()

```

##Flatten numpy array for easy training with KNN

```python
X_train.shape
```

```python
X_train=X_train.reshape(60000, 28*28)
```

```python
X_train[1].shape
```

#Train model using K Nearest Neighbor Model Classifier





```python
#Import K Nearest Neighbor class from sklearn
from sklearn.neighbors import KNeighborsClassifier
```

```python
model = KNeighborsClassifier(n_neighbors=3)
model.fit(X_train, y_train)
```

#Evaluate Model

```python
#import needed modules
from sklearn.metrics import accuracy_score, classification_report
```

##Flatten test dataset 

```python
X_test=X_test.reshape(10000,28*28)
```

```python
predictions=model.predict(X_test)
```

```python
accuracy=accuracy_score(predictions, y_test)
print("Accuracy score: ", accuracy)
```

```python
report = classification_report (predictions,y_test)
print(report)
```
